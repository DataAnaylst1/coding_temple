### **Task 1: Update Shopping Cart**
Using the shopping cart you created yesterday as a blueprint, create a shopping cart using object-oriented programming. This shopping cart should maintain all functionality that you currently have in your blueprint.

```python
# YOUR CODE HERE
class ShoppingCart:
    """
    The ShoppingCart will have items, capacity, available items, and total

    Attributes for this class are:
    - items: expected to be a list
    - capacity: expected to be a string or integer
    - available_items: expected to be a string

    Methods of this class are:
    - add_items_with_quantity_and_price_total: add an item and it's quanity to the cart based on user's input
    - display_shopping_cart: display items in case
    - remove_items: remove items from cart based on user's request
    """

    def __init__(self, items: list, capacity):
        self.items = items
        self.capacity = capacity
    
    def display_shopping_cart(self):
        """Displays current items in shopping bag"""
        print("Here are the items that are in your shopping bag!")
        # Prints each item in the shopping bag
        for item in self.items():
            print(item)
        
    def add_items_quanity_and_price_total(self):
        """Add item to shopping cart based on user's input"""
        product = input("Please highlight which item you would like to add to your shopping cart: \n\t Mac Laptop: $1,000 \n PS5: $543.99 \n Living Room Set: $1,500.00")
        quantity = (input("Great choice! How many?"))
        if quantity.isdigit():
            quantity = int(quantity)
        else:
            quantity = input("Please provide an actual number.\nHow many would you like?")
            quantity = int(quantity)
        product_and_quantity = {}
        if product.lower() == "mac laptop":
            price = 1_000
            total = price*quantity
        if product.lower() == "ps5":
            price = 543.99
            total = price*quantity
        if product.lower() == "living room set":
            price = 1_500
            total = price*quantity
        product_and_quantity[product]=[total]
        self.items.append(product_and_quantity)


    def remove_items(self):
        """Remove items based on user input"""
        for item in self.items():
            print(item)
            remove_specific_item = input("Which item would you like to remove?")
            for i in range(len(remove_specific_item)):
                if remove_specific_item.lower()[i] != self.items[i]:
                    print("This item is not found in your shopping bag")
                else:
                    del self.items[remove_specific_item]
    

        
```
