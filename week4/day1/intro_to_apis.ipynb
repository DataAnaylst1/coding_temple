```python
import requests
import pandas as pd
from PIL import Image
from io import BytesIO

api_url = 'https://pokeapi.co/api/v2/pokemon/1/'
response = requests.get(api_url)
response.json()

```

```python
"""Pokemon #1 Bulbasaur"""
name = response.json()['forms'][0]['name'] # The name stored
ability = response.json()['abilities'][0]['ability']['name'] # An ability name stored
base_experience = response.json()['base_experience'] # The base experience stored
url = response.json()['forms'][0]['url'] # The URL stored
img_str = requests.get(response.json()['forms'][0]['url']) # The URL stored
img_str.json()['sprites']['front_shiny'] # Link to the front shiny image stored
# Image.open(BytesIO(img_str.content))

"""Create pokemon dictionary """
pokemon_dict={
            'name': [name],
            'ability': [ability],
            'base_experience': [base_experience],
            'url': [url]
        }
pokemon_dict
"""Convert into Dataframe"""
pd.DataFrame.from_dict(pokemon_dict)



```
