```python
#Which neighorhood group has the highest average price?

# result = air_bnb.groupby(['neighbourhood_group']).mean(numeric_only = True).reset_index(name='count')
# We can use .mean(numeric_only = True) if you want to include another column that is not numeric
# result = air_bnb.groupby('neighbourhood_group')['price'].mean()
#Which neighorhood group has the highest average price?

# result = air_bnb.groupby('neighbourhood_group')['price'].mean()
# result2 = result.round(2)
# display(result2)
# ANOTHER WAY BELOW

# result = air_bnb.groupby('neighbourhood_group')['price'].mean().sort_values('mean price', ascending=False)
# result2 = result.round(2)
# display(result2)

# result2 = result.loc[result['number_of_reviews'] == 629]

result = air_bnb.groupby('neighbourhood_group')['price'].mean().round(2).sort_values(ascending = False).reset_index(name='mean price')
display(result)

# Answer:
# Manhatten
```

```python
# Which neighbor hood group has the highest total price?

# count= display(air_bnb['price'].describe())
# count

result = air_bnb.groupby('neighbourhood_group')['price'].sum().sort_values(ascending = False).reset_index(name='total price')
result
```

```python
#Which top 5 hosts have the highest total price?

# air_bnb.head()
result = air_bnb.groupby('host_name')['price'].mean().round(2).sort_values(ascending = False).reset_index(name='mean price')
result.head()



```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

df = air_bnb[(air_bnb['number_of_reviews'] >=100) & (air_bnb['availability_365'] == 0)]
df

results = df.sort_values('number_of_reviews', ascending=False).groupby(['host_name','availability_365']).head()
display(results)

# result = result['number_of_reviews'] >=100 and result['availability_365'] == 0
# display(air_bnb[results])
```

```python
# What host has the highest total of prices and where are they located?

result = air_bnb.groupby(['host_name', 'neighbourhood_group'])['price'].sum().round(2).sort_values(ascending=False).reset_index(name='total prices').head(10)
display(result)
```

```python
# When did Danielle from Queens last receive a review?

danielle = air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')]
result2 = danielle.sort_values('last_review', ascending=False).reset_index(drop=True).head(1)
display(result2)



```
